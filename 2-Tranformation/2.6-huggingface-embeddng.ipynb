```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
embeddings
```

```python
text= "The quick brown fox jumps over the lazy dog. The quick brown fox jumps over the lazy dog. The quick brown fox jumps over the lazy dog."
results = embeddings.embed_query(text)
results
```

```python
len(results)  # Length of the embedding vector
```

```python
# for documents
docs = ["The quick brown fox jumps over the lazy dog.",
        "jimbo is the killer"]
results2 = embeddings.embed_documents(docs)
results2
```
