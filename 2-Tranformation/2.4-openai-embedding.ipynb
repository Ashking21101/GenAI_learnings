converting text to vectors


```python
import os
from dotenv import load_dotenv
load_dotenv()# to load environment variables from .env file
```

```python
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")  # Set your OpenAI API key
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model = "text-embedding-3-large")
embeddings 
```

```python
text = "This is a sample text to be embedded."
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)  # Length of the embedding vector
```

```python
query_result[0]
```

I can also set my own dimension, unlike above which is 3072 etc

```python
embeddings_1024 = OpenAIEmbeddings(model = "text-embedding-3-large", dimensions=1024)
query_result_1024 = embeddings_1024.embed_query(text)
query_result_1024
```

```python
len(query_result_1024)
```

```python
query_result_1024[0]
```

# now in this step we will do as following
use textloader

use RecurssiveTextSplitter

```python
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

loader = TextLoader(r'speeh.txt')
docs = loader.load()

docs
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 30, chunk_overlap=3)
final_docs=text_splitter.split_documents(docs)
final_docs
```

# Vector ebedding and Vector DB (for storing)

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_docs, embeddings_1024)
# 1st argument is the splitted documents, 2nd is the embeddings
db
```

# how to query from this vector DB

```python
# so using the following text 
# # we are finding the similarity of it from the entire text 
# that is in the vector DB and we are getting the most similar text to the query text

query = "This is our hope. This is the faith that I go back to the South with."
retrieved_results = db.similarity_search(query)
print(retrieved_results)
```
